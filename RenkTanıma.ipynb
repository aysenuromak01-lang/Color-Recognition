```python
import pandas as pd
import numpy as np
```

```python
index = ["color_code", "color_name", "hex", "R", "G", "B"]
df = pd.read_csv('colors.csv', names=index, header=None)
# 1. Renk veri setini doğrudan URL üzerinden yükleyin
#csv_url = "https://github.com/amankharwal/Website-data/blob/master/colors.csv"
```

```python
# 2. En Yakın Renk İsmini Bulan Fonksiyon (Mesafe Hesaplama)
def get_color_name(R, G, B):
    minimum = 10000
    color_name = ""
    for i in range(len(df)):
        # Tıklanan RGB ile veri setindeki renklerin RGB farkını hesapla
        d = abs(R - int(df.loc[i, "R"])) + abs(G - int(df.loc[i, "G"])) + abs(B - int(df.loc[i, "B"]))
        if d <= minimum:
            minimum = d
            color_name = df.loc[i, "color_name"]
    # O satırdaki renk adını döndürün
    return df.loc[min_index, 'color_name']
```
